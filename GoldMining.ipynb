```python
import pandas as pd
gold = pd.read_csv("gold.csv")
dollars = pd.read_csv("dollars.csv")
oil = pd.read_csv("oil.csv")
# missing value preprocessing
gold.dropna(inplace = True)
gold.isnull().sum()
```

```python
dollars.dropna(inplace = True)
dollars.isnull().sum()
```

```python
oil.dropna(inplace = True)
oil.isnull().sum()
```

```python
import numpy as np
from sklearn import preprocessing
#data normalization
goldn = preprocessing.normalize(np.array(gold.Price).reshape(1,-1), norm='l1')
goldn = goldn.flatten()
goldn
```

```python
oiln = preprocessing.normalize(np.array(oil.Price).reshape(1,-1), norm="l1")
oiln = oiln.flatten()
oiln
```

```python
dollarsn = preprocessing.normalize(np.array(dollars.Price).reshape(1,-1), norm="l1")
dollarsn = dollarsn.flatten()
dollarsn
```

```python
from datetime import datetime
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
date = np.array(gold.Time)
datex = [datetime.strptime(d, '%Y/%m/%d').date() for d in date]
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y/%m/%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
datex

date = np.array(dollars.Time)
datey = [datetime.strptime(d, '%Y/%m/%d').date() for d in date]
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y/%m/%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
datey

date = np.array(oil.Time)
datez = [datetime.strptime(d, '%Y/%m/%d').date() for d in date]
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y/%m/%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
datez

plt.figure()
plt.plot(datex, goldn, color='green',linewidth=4)
plt.plot(datey, dollarsn, color='red',linewidth=4)
plt.plot(datez, oiln, color='blue',linewidth=4)
plt.gcf().autofmt_xdate()
plt.show()
```
